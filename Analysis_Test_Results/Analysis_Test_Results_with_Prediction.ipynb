## Trying to train the data 

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
features = test_report[['Total_Sent', 'Sent_On_Date', 'Proceed_In_24_Hr', 'Total_Positive']]
label = test_report[['New_Positive']]
```

```python
features
```

```python
label
```

```python
# Create a correlation matrix
corr_metrics = features.corr()
corr_metrics.style.background_gradient()
```

## Finding the important features 

```python
import rfpimp
from sklearn.ensemble import RandomForestRegressor
```

```python
x_train,x_test,y_train,y_test = train_test_split(features,label,test_size = 0.3,random_state=42)
```

```python
rf = RandomForestRegressor(n_estimators=100,n_jobs=-1)
rf.fit(x_train,y_train)
```

```python
imp = rfpimp.importances(rf,x_test,y_test)
```

```python
fig,ax = plt.subplots(figsize=(10,5))
ax.barh(imp.index,imp['Importance'],height=0.8,alpha=0.8)
ax.set_xlabel('Importance score')
ax.set_title('Permutation feature importance')
plt.gca().invert_yaxis()
fig.tight_layout()
```

```python
x_trainsample =x_train[['Proceed_In_24_Hr','Total_sent','Total_positive']]
x_testsample = x_test[['Proceed_In_24_Hr','Total_sent','Total_positive']]
```

```python
model = LinearRegression()
```

```python
model.fit(x_trainsample,y_train)
```

```python
predicted = model.predict(x_testsample)
```

```python
model.score(x_testsample,y_test)
```

```python
model.score(x_trainsample,y_train)
```

```python
error = metrics.mean_squared_error(y_test,predicted)
print('root mean squared error : ',np.sqrt(error))
r2_Score = metrics.r2_score(y_test,predicted)
print("r2 score : ",r2_Score)
```
