### feature selection 

```python
# Defining a function to evaluate a model
def evaluateModel(model):
    print("RSS = ", ((test_report.New_Positive - model.predict())**2).sum())
    print("R2 = ", model.rsquared)
```

```python
# for Total_Sent
model_Total_Sent  = sm.ols('New_Positive ~ Total_Sent',test_report).fit()
evaluateModel(model_Total_Sent)
```

```python
# for Sent_On_Date
model_Sent_On_Date  = sm.ols('New_Positive ~ Sent_On_Date',test_report).fit()
evaluateModel(model_Sent_On_Date)
```

```python
# for Proceed_In_24_Hr
model_Proceed_In_24_Hr  = sm.ols('New_Positive ~ Proceed_In_24_Hr',test_report).fit()
evaluateModel(model_Proceed_In_24_Hr)
```

```python
# for Total_Positive
model_Total_Positive  = sm.ols('New_Positive ~ Total_Positive',test_report).fit()
evaluateModel(model_Total_Positive)
```

```python
#for total positive processed in 24 
model_Total_Positive_processed24  = sm.ols('New_Positive ~ Total_Positive + Proceed_In_24_Hr',test_report).fit()
evaluateModel(model_Total_Positive_processed24)
```

```python
#for total positive processed in 24 Total_Sent
model_Totalpos_processed24sent  = sm.ols('New_Positive ~ Total_Positive + Proceed_In_24_Hr + Total_Sent',test_report).fit()
evaluateModel(model_Totalpos_processed24sent)
```

```python
#for total positive processed in 24 sendondate
model_Totalpos_processed24_ondate  = sm.ols('New_Positive ~ Total_Positive + Proceed_In_24_Hr + Sent_On_Date',test_report).fit()
evaluateModel(model_Totalpos_processed24_ondate)
```

```python
#while combining all
model_all = sm.ols('New_Positive ~ Total_Positive + Proceed_In_24_Hr + Sent_On_Date + Total_Sent ',test_report).fit()
evaluateModel(model_all)
```

```python
model_Total_Positive_processed24  = sm.ols('New_Positive ~ Total_Positive + Proceed_In_24_Hr',test_report)
resuts_formula = model_Total_Positive_processed24.fit()
x_surf,y_surf = np.meshgrid(np.linspace(test_report.Total_Positive.min(),test_report.Total_Positive.max(),100),np.linspace(test_report.Proceed_In_24_Hr.min(),test_report.Proceed_In_24_Hr.max(),100))
onlyX = pd.DataFrame({'Total_Positive':x_surf.ravel(),'Proceed_In_24_Hr':y_surf.ravel()})
fittedY = resuts_formula.predict(exog = onlyX)
fittedY = np.array(fittedY)
```
